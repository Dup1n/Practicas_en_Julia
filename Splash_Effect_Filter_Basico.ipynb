# Practica 2.- Splash Effect Filter de una Imagen (Basico)
### Problema de procesamiento de imagenes, sobre el filtro de colores en una region de la imagen 
Nos descargaremos el Paquete de Julia dedicado a imagenes `Pkg.add("Images")` y lo utilizamos mediante el siguiente comando:

```julia
using Images
```

Cargamos una imagen:

```julia
img = load("data/kurisu.png")
```

Seleccionamos una region de la imagen la cual se usara como filtro: 

```julia
region_filter = img[220:252,992:end] #Region del Cabello
```

Creamos una funcion `colour_splash_filter(pixel, region)` que actua en cada pixel y compara con la region tomada para, si no esta en ella el pixel sera transformado en una Escala de Gris, si no, se man

```julia
function colour_splash_filter(pixel, region_filter)
    is_color = true
    for color in region_filter
        if pixel == color
            is_color = true
            break
        end
        is_color = false
    end
    if !is_color
        pixel = RGBA.(Gray.(pixel))
    end
    pixel
end
```

Mapeamos la imagen con su respectiva funcion y la region para obtener el resultado

```julia
img_filter = map(pixel -> colour_splash_filter(pixel, region_filter), img)
```

Probando con otra region de la imagen:

```julia
region_filter_h = img[1:end,992:end] #Ultimas hileras de la imagen
```

Mapeamos la imagen con su respectiva funcion y la region para obtener el resultado

```julia
img_filter_h = map(pixel -> colour_splash_filter(pixel, region_filter_h), img)
```

Mostramos en conjunto las imagenes: 
    - Imagen Normal (RGB) | Imagen Filtro (Cabello) | Imagen Filtro (Hilera)

```julia
mosaicview(img, img_filter, img_filter_h; nrow = 1, npad=10)
```
