# Practica 1.- Color inverso de una Imagen
### Empezaremos con un problema sencillo de procesamiento de imagenes, el calculo de la inversa mediante Escala de Grises y RGB en Julia
Nos descargaremos el Paquete de Julia dedicado a imagenes `Pkg.add("Images")` y lo utilizamos mediante el siguiente comando:

```julia
using Images
```

Cargamos una imagen:

```julia
img = load("data/kurisu.png")
```

La pasamos a Escala de Grises:

```julia
imgG = Gray.(img)
```

Creamos dos funciones que actuaran en cada pixel de las imagenes, debemos tener en cuenta que una imagen usualmente usa RGB para el color de estas, lo que hace que cada pixel esta compuesto con atributos `pixel.r pixel.g pixel.b` mientras que Escala de Grises solo posee un valor, ambas funciones poseen la misma estructura

```julia
function negative_image_rgb(img_pixel)
    img_pixel = RGB(1 - img_pixel.r, 1 - img_pixel.g, 1 - img_pixel.b)
end

function negative_image_gray(img_pixel)
    img_pixel = Gray(1 - img_pixel)
end
```

Mapeamos las imagenes con su respectiva funcion para obtener el resultado

```julia
img_negative_rgb = map(img_pixel -> negative_image_rgb(img_pixel), img);
img_negative_gray = map(img_pixel -> negative_image_gray(img_pixel), imgG);
```

Mostramos en conjunto las imagenes: 
    - Imagen Normal (RGB) | Imagen Inverso (RGB)
    - Imagen Normal (Grises) | Imagen Inverso (Grises)

```julia
mosaicview(img, imgG, img_negative_rgb, img_negative_gray; ncol = 2, npad=10)
```
